## The aim of the below code is to summarize text using google pegasus-xsum which is a pre-trained deep-learning model implemented using Hugging phase transformers.The Pegasus model is built using a Transformer Encoder-Decoder architecture which is widely used for text summarization.

```python
#Installing transformers.
!pip install transformers
```

```python
#intialising a sentence that has to be summarised
sentence_to_summarise = """The tower is 324 metres (1,063 ft) tall, about the same height as an 81-storey building, and the tallest structure in Paris. Its base is square, measuring 125 metres (410 ft) on each side. During its construction, the Eiffel Tower surpassed the Washington Monument to become the tallest man-made structure in the world, a title it held for 41 years until the Chrysler Building in New York City was finished in 1930. It was the first structure to reach a height of 300 metres. Due to the addition of a broadcasting aerial at the top of the tower in 1957, it is now taller than the Chrysler Building by 5.2 metres (17 ft). Excluding transmitters, the Eiffel Tower is the second tallest free-standing structure in France after the Millau Viaduct."""
#intialising the tranaformers pipeline
from transformers import pipeline
# using the pegasus-xsum for text summarization
classifier = pipeline("summarization",model="google/pegasus-xsum")
print(classifier(sentence_to_summarise))
```
